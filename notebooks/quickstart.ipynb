```python
# ============================================================
# BSA/AML Regulatory RAG
# AI for Fintech | [GENAI-001]
# ============================================================

import os
import sys
from dotenv import load_dotenv

sys.path.append(os.path.abspath('..'))
load_dotenv()
```

```python
# Pipeline Initialization
# Loads the document, processes chunks, builds vector store and LLM

from src.ingestion import load_pdf
from src.processing import process_documents
from src.embedding import build_vector_store
from src.retrieval import build_bm25
from src.generation import build_llm

extracted_pages = load_pdf()
document_chunks = process_documents(extracted_pages)
chroma_collection, embedding_model = build_vector_store(document_chunks)
bm25_index = build_bm25(document_chunks)
llm = build_llm()

print('Pipeline ready')
```

```python
# Ask any question about BSA/AML compliance
# Retrieves relevant chunks via hybrid search and generates a cited answer
from src.retrieval import hybrid_search'
from src.generation import generate_answer

def run(user_query):
    '''
    Runs the full RAG pipeline for a given query.
    Retrieves relevant chunks and returns a cited answer.
    '''
    
    retrieved_chunks = hybrid_search(
        user_query, 
        chroma_collection, 
        embedding_model, 
        bm25_index, 
        document_chunks
    )

    answer = generate_answer(user_query, retrieved_chunks, llm)
    print(f'\nQuery: {user_query}')
    print(f'\nAnswer:\n{answer}')
    
    return answer
```

```python
question = ''' 
    What are the internal controls required for BSA/AML compliance?
'''
run(question)
```
